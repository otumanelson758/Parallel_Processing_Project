````python
# Your code goes here
def grep_mapper(file_chunk):
    """Retrieves the line index of a target word in a file

    Args:
      file_chunk (iterable of string): A file folder

    Returns:
       dictionary {str:list}
    """
    match_result = {}
    folder_name = "/content/drive/MyDrive/Colab Notebooks/Data Engineering/wiki"
    for i in range (len(file_chunk)):
        file_name = file_chunk[i]
        word_index = []
        with open(os.path.join(folder_name, file_name),encoding = "ISO-8859-1") as f:
            lines = [line for line in f.readlines()]
            chunk_lines = list(map(lambda x: x.split(), lines))
        for j in range(len(chunk_lines)):
          line = chunk_lines[j]
          for word in line:
            if word == target:
              word_index.append(j)
              match_result[file_name] = word_index
    return match_result
````

````python
#We do a mapper function to fix this.
def grep_reducer(match_result1, match_result2):
    """Merges two dictionaries

    Args:
      match_result1 {str:list}
      match_result2 {str:list}

    Returns:
       dictionary {str:list}
    """
    merged_results = {}
    merged_results.update(match_result1)
    merged_results.update(match_result2)
    return merged_results
# We need to call our files and save them 'file_names'
````

## 4. Grep Case Insensitive

Let's improve our grep function by making it case insensitive. This means that the case of the characters in the strings won't matter.

For example, imagine that there are two files with the following contents:

Screenshot 2021-11-10 at 08.47.26.png

We want to modify the function so that if we search again for the string "more", we would find the same three matches as before:

Screenshot 2021-11-10 at 08.47.31.png


**Tasks**

* Use MapReduce to create a function that, given a string, creates a dictionary there. The keys are the filenames, and the values are lists with all line indices that contain the given string.

* Use the function to find all occurrences of the string "data" in the files stored in the wiki folder.



````python
# Your code goes here
import re
def grep_mapper_allcase(file_chunk):
    """Retrieves the line index of a target word in a file

    Args:
      file_chunk (iterable of string): A file folder

    Returns:
       dictionary {str:list}
    """
    #we create and empty list called match_result.
    #we still pass our folder wiki dataset to folder_name
    match_result = {}
    folder_name = "/content/drive/MyDrive/Colab Notebooks/Data Engineering/wiki"
    for i in range (len(file_chunk)):
        file_name = file_chunk[i]
        word_index = []
        with open(os.path.join(folder_name, file_name),encoding = "ISO-8859-1") as f:
            lines = [line for line in f.readlines()]
            chunk_lines = list(map(lambda x: x.split(), lines))
        for j in range(len(chunk_lines)):
          line = chunk_lines[j]
          for word in line:
            word = re.sub(r"[^a-z0-9]","",word.lower())
            if word == target:
              word_index.append(j)
              match_result[file_name] = word_index
    return match_result
    #we return the match_result 
````

## 6. Finding Match Positions on Lines

Right now, we are only finding the line numbers where there is at least one occurrence. Let's extend the algorithm so that it provides information about the location of the matches in those lines.

The following diagram shows one line of the Data Science Wikipedia page. For the sake of this example, let's imagine that this line has index 3.

The line contains three occurrences of the string "data" at indexes 0, 27, and 65 (ignoring case):

Screenshot 2021-11-10 at 08.53.05.png

The current implementation will just return the index of this line: 3. The new implementation should return pairs of indices where the first value is the line index and the second index if the index of the first character of the match on that line. For the above example, the output for this line would be: (3, 0), (3, 27) and (3, 65).

**Task**

* Modify the mapper function so that it returns a list of pairs with all occurrences of the target on all files from the given chunk. In each pair, the first number should be the line index and the second number should be the index on the line.



````python
# Your code goes here
#Modify the mapper function so that it returns a list of pairs with all 
#occurrences of the target on all files from the given chunk.
def grep_mapper_matchpos(file_chunk):
    """Retrieves the word index of a target word in a file
    and a summary of the line

    Args:
      file_chunk (iterable of string): A file folder

    Returns:
       dictionary {str:tuple(int,int,str)}
    """
    # create an empty list and assign it a variable match_result
    #folder name points to the file location
    match_result = {}
    folder_name = "/content/drive/MyDrive/Colab Notebooks/Data Engineering/wiki"
    for i in range (len(file_chunk)):
        file_name = file_chunk[i]
        #we will still iterate through the lines in chunks
        word_index = ()
        with open(os.path.join(folder_name, file_name),encoding = "ISO-8859-1") as f:
            lines = [line for line in f.readlines()]
            chunk_lines = list(map(lambda x: x.split(), lines))
        for j in range(len(chunk_lines)):
#each pair, the first number should be the line index and the second number should be the index on the line.         
          line = chunk_lines[j]
          line_ = ' '.join(map(str, line))
          line__ = re.sub(r"[^a-z0-9]", " ", line_.lower())
          for word in line__.split():
            if word == target:
              word_index = j,line__.find(target),line_
              match_result[file_name] = word_index
    return match_result
    #return the match result.
````

# Each pair, the first number should be the line index and the second number should be the index on the line.
